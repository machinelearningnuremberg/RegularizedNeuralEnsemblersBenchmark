```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_sizes = pd.read_csv("saved_results/dataset_sizes.csv")
dataset_sizes.columns = ["index", "dataset_complete_id", "dataset_name", "dataset_size", "num_of_classes"]
```

```python
results1 = pd.read_csv("saved_results/Neural_32hd_random.csv")
results2 = pd.read_csv("saved_results/Random.csv")
```

```python
results1_agg = pd.merge(results1, dataset_sizes, on="dataset_complete_id", how="inner")
results2_agg = pd.merge(results2, dataset_sizes, on="dataset_complete_id", how="inner")
```

```python
results1_agg.columns
```

```python
sns.scatterplot(results1_agg, y='incumbent_ensemble_test_metric', x="dataset_size", label="1")
sns.scatterplot(results2_agg, y='incumbent_ensemble_test_metric', x="dataset_size", label="2")
plt.legend()
```

```python
sns.scatterplot(results1_agg, y='incumbent_ensemble_test_metric', x="num_of_classes", label="1")
sns.scatterplot(results2_agg, y='incumbent_ensemble_test_metric', x="num_of_classes", label="2")
plt.legend()
```
